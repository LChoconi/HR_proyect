```python
def create_server_connection(host_name, user_name, user_password):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
def create_db_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name,
            allow_local_infile = 1 # Necessary to load data
        )
        print("MySQL Database connection successful")
    except Error as err:
        print(f"Error: '{err}'")

    return connection
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query successful")
    except Error as err:
        print(f"Error: '{err}'")
```

### Create the calendar table with the desired projection period

```python
create_calendar_table = """
CREATE TABLE calendar AS
WITH RECURSIVE CTE AS (
	-- customize start date here
	select date('2023-09-01') as calendar_date
	union all
	select date_add(calendar_date, interval 1 day) as calendar_date from cte 
	-- customize end date here
	where (date_add(calendar_date, interval 1 day)) <= date('2024-12-31')
)
SELECT DISTINCT
YEAR(calendar_date) as year,
MONTHNAME(calendar_date) as month,
LAST_DAY(calendar_date) as date
FROM cte;
 """

execute_query(connection, create_calendar_table) # Execute our defined query
```

### UNION Headcount & Movements

```python
create_union_table = """
CREATE TABLE union_hc_mov AS
SELECT 
EmpID AS ID,
Position,
Department,
quantity,
Salary,
monthly_salary,
year,
month,
date,
date_mov,
base
FROM activeheadcount_projected
UNION ALL
SELECT
MovID AS ID,
Position,
Department,
Quantity,
salary_x_Q AS Salary,
monthly_salary,
year,
month,
date,
dateofmov_f AS date_mov,
base
FROM wfmovements_projected;
 """

execute_query(connection, create_union_table)
```

### Create and Populate salaryIncrease Table.

```python
create_table = """
CREATE TABLE salaryIncrease (
    id INT,
    year INT,
    month VARCHAR(10),
    date DATE,
    increase_percentaje DECIMAL(10,5)
  );
 """

populate_table = """
LOAD DATA LOCAL INFILE  
'./salaryIncrease.csv'
INTO TABLE salaryIncrease  
COLUMNS TERMINATED BY ';'
LINES TERMINATED BY '\r\n'
IGNORE 1 LINES
(id, year,
    month,
    date,
    increase_percentaje);
 """

execute_query(connection, create_table)
execute_query(connection, populate_table) 
```

```python
create_table = """
CREATE TABLE salaryincrease_index AS
WITH RECURSIVE cte AS (
  SELECT id, year, month, date, increase_percentaje, 1*(1+ increase_percentaje/100) AS increase_index, id-1 AS count_with
  FROM salaryincrease
  UNION ALL
  SELECT t.id, t.year, t.month, t.date, t.increase_percentaje, increase_index * (1+t.increase_percentaje/100), count_with
  FROM salaryincrease t
  JOIN cte c
    ON c.id = t.id-1
)
SELECT DISTINCT
	id, year, month, date, increase_percentaje, increase_index
FROM cte
WHERE count_with=0
ORDER BY id, date;
 """

execute_query(connection, create_table)
```

### Create the budget table and calculate the value for each month based in the increase index.

```python
create_table = """
CREATE TABLE budget_labor_Cost AS
(
SELECT
	L.ID,
	L.Position,
    L.Department,
    L.quantity,
    L.Salary,
    L.year,
    L.month,
    L.date_mov,
    L.base,
    R.increase_index
FROM 
	union_hc_mov L
LEFT JOIN salaryincrease_index R
	USING (date)

)
 """

execute_query(connection, create_table)
```

```python
create_columns_WFM1 = """
ALTER TABLE budget_labor_Cost
ADD budget_value DECIMAL(30,5);
"""

create_columns_WFM2 = """
UPDATE  budget_labor_Cost
SET budget_value = Salary/12*increase_index;
"""


execute_query(connection, create_columns_WFM1)
execute_query(connection, create_columns_WFM2)
```
